```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
d=pd.read_csv("C:\\Users\\91977\\mall_customers.csv")
d.head()
X=d.iloc[:,:].values
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.xlabel("customers")
plt.ylabel("eclidean distance")
plt.show()
                          
```

```python
from sklearn.cluster import AgglomerativeClustering
clustering = AgglomerativeClustering(n_clusters=5)
train_hc=clustering.fit_predict(X)
```

```python
train_hc
```

```python
plt.scatter(X[train_hc == 0,0],X[train_hc == 0,1],c="red",label="c1")
plt.scatter(X[train_hc == 1,0],X[train_hc == 1,1],c="yellow",label="c2")
plt.scatter(X[train_hc == 2,0],X[train_hc == 2,1],c="blue",label="c3")
plt.scatter(X[train_hc == 3,0],X[train_hc == 3,1],c="green",label="c4")
plt.scatter(X[train_hc == 4,0],X[train_hc == 4,1],c="violet",label="c5")
plt.xlabel("annual income")
plt.ylabel("spending score")
plt.legend()
plt.show()
```
